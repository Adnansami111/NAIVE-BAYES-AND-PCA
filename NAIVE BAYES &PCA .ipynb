```python
#gaussianNB
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB# IT IS USED FOR CONTINOUS INPUTS LIKE DECIMAL TO PREDICT CATECORICAL WITH COMBINATION OF STANDARD SCALER 
from sklearn.metrics import accuracy_score,precision_score,recall_score# here with this performance metrices
```

```python
#to find outliers
n=[ 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
for col in n:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
n=[ 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
for col in n:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
df=df.drop(columns=["Unnamed: 32","id"])
df
df.columns
```

```python
X=df.drop(columns=["diagnosis"],axis=1)
Y=df["diagnosis"]
```

```python
Y=pd.get_dummies(Y)["M"]#encoding is done for output(target value) since it is only categorical values.
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)#it wil split dataset into train test as 80 20
```

```python
scaler=StandardScaler()
X_train_scale=scaler.fit_transform(X_train)# it perform aggeressive
X_test_scale=scaler.transform(X_test)
```

```python
model=GaussianNB(var_smoothing=1e-9,priors=None)# VAR SMOOTHING IS USED TO VARIENCE IS LOWER and priors is probality of target
model.fit(X_train_scale, Y_train)
```

```python
Y_pred=model.predict(X_test_scale)
```

```python
print(accuracy_score(Y_test,Y_pred))
print(precision_score(Y_test,Y_pred))
print(recall_score(Y_test,Y_pred))
```

```python
#MULTINOMIALNB
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer # it is used for text dataset 

from sklearn.naive_bayes import MultinomialNB# it is used spam detection only or text data with combination count vectorizer
from sklearn.metrics import accuracy_score,precision_score,recall_score# here with this performance metrices
```

```python
df=pd.read_csv("spam mail.csv")
```

```python
df
```

```python
df.columns
```

```python
df['Category'].map({'ham':0,'spam':1})
```

```python
df["Masseges"]
```

```python
vectorizer=CountVectorizer(binary=True)
```

```python
X=vectorizer.fit_transform(df["Masseges"])
Y=df["Category"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model=MultinomialNB(alpha=1.0,fit_prior=True) # Alpha is used to increase fertures count and numerical stability promote fit prior  is used for 
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
print(accuracy_score(Y_test,Y_pred))
```
